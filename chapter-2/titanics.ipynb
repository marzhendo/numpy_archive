```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data penumpang titanic
# [PassengerId, Survived, Class, Age]
# Survived: 0 = No, 1 = Yes
# Class: 1 = Eksekutif, 2 = Bisnis, 3 = Ekonomi
passengers = np.array([
   [1, 0, 3, 22],
   [2, 1, 1, 38],
   [3, 1, 3, 26],
   [4, 1, 1, 35],
   [5, 0, 3, 35],
   [6, 0, 3, 18],
   [7, 0, 1, 54],
   [8, 0, 3, 2],
   [9, 1, 3, 27],
  [10, 1, 2, 14],
  [11, 1, 3, 4],
  [12, 1, 1, 58],
  [13, 0, 3, 20],
  [14, 0, 3, 39],
  [15, 0, 3, 14],
  [16, 1, 2, 55],
  [17, 0, 3, 2],
  [18, 1, 2, 12],
  [19, 0, 3, 31],
  [20, 1, 3, 8],
  [21, 0, 2, 35],
  [22, 1, 2, 34],
  [23, 1, 3, 15],
  [24, 1, 1, 28],
  [25, 0, 3, 8],
  [26, 1, 3, 38],
  [27, 0, 3, 2],
  [28, 0, 1, 1],
  [29, 1, 3, 5],
  [30, 0, 3, 18],
  [31, 0, 1, 40],
  [32, 1, 1, 70],
  [33, 1, 3, 33],
  [34, 0, 2, 66],
  [35, 0, 1, 28],
  [36, 0, 1, 42],
  [37, 1, 3, 5],
  [38, 0, 3, 18],
  [39, 0, 3, 18],
  [40, 1, 3, 14],
  [41, 0, 3, 40],
  [42, 0, 2, 27],
  [43, 0, 3, 29],
  [44, 1, 2, 0],
  [45, 1, 3, 19],
  [46, 0, 3, 33],
  [47, 0, 3, 14],
  [48, 1, 3, 22],
  [49, 0, 3, 41],
  [50, 0, 3, 18]
])
# Cek dimensi array passengers
print("Dimensi array passengers:", passengers.shape)
# Rata-rata umur penumpang
avg_age = np.mean(passengers[:, 3])
print("Rata-rata umur penumpang:", avg_age)
# Index penumpang dengan umur tertua
oldest_index = np.argmax(passengers[:, 3])
print("Index penumpang dengan umur tertua:", oldest_index)
# Index penumpang dengan umur termuda
youngest_index = np.argmin(passengers[:, 3])
print("Index penumpang dengan umur termuda:", youngest_index)
# Ambil ID dari penumpang tertua dan termuda
oldest_id = passengers[oldest_index, 0]
youngest_id = passengers[youngest_index, 0]
print("ID penumpang tertua:", oldest_id)
print("ID penumpang termuda:", youngest_id)
# Hitung persentase penumpang yang selamat
survival_rate = np.mean(passengers[:, 1])
print(f'Presentase penumpang yang selamat: {survival_rate * 100:.2f}%')
# “Berapa persentase penumpang yang selamat berdasarkan kelas penumpang mereka?”
# Ambil data penumpang yang berasal dari kelas 1
class_1_passengers = passengers[passengers[:, 2] == 1]
# Hitung persentase penumpang kelas 1 yang selamat
class_1_survival_rate = np.mean(class_1_passengers[:, 1])
# Ambil data penumpang yang berasal dari kelas 2
class_2_passengers = passengers[passengers[:, 2] == 2]
# Hitung persentase penumpang kelas 2 yang selamat
class_2_survival_rate = np.mean(class_2_passengers[:, 1])
# Ambil data penumpang yang berasal dari kelas 3
class_3_passengers = passengers[passengers[:, 2] == 3]
# Hitung persentase penumpang kelas 3 yang selamat
class_3_survival_rate = np.mean(class_3_passengers[:, 1])
print(f'Presentase penumpang kelas 1 yang selamat: {class_1_survival_rate * 100:.2f}%')
print(f'Presentase penumpang kelas 2 yang selamat: {class_2_survival_rate * 100:.2f}%')
print(f'Presentase penumpang kelas 3 yang selamat: {class_3_survival_rate * 100:.2f}%')


# Visualisasi data penumpang titanic berdasarkan umur dan status keselamatan
survived = passengers[passengers[:, 1] == 1]
not_survived = passengers[passengers[:, 1] == 0]
plt.scatter(survived[:, 0], survived[:, 3], color='green', label='Selamat')
plt.scatter(not_survived[:, 0], not_survived[:, 3], color='red', label='Tidak Selamat')
plt.title("Data Penumpang Titanic berdasarkan Umur dan Status Keselamatan")
plt.xlabel("Passenger ID")
plt.ylabel("Umur")
plt.legend()
plt.show()
```
