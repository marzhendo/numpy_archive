```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
steps = np.array([4500, 6200, 3100, 8500, 5000, 2500, 7800])
# Langkah paling minimun dalam seminggu
min_steps = np.min(steps)
print("Langkah paling minimum dalam seminggu:", min_steps)
# Langkah paling maksimum dalam seminggu
max_steps = np.max(steps)  
print("Langkah paling maksimum dalam seminggu:", max_steps)
# Rata-rata langkah dalam seminggu
avg_steps = np.mean(steps)  
print("Rata-rata langkah dalam seminggu:", avg_steps)
# Total langkah dalam seminggu
total_steps = np.sum(steps)
print("Total langkah dalam seminggu:", total_steps)
```

```python
data_langkah = np.array([
    [4500, 6200, 3100, 8500, 5000, 2500, 7800], # Kamu
    [2000, 3000, 4000, 5000, 6000, 7000, 8000], # Adik
    [9000, 8000, 7000, 6000, 5000, 4000, 3000]  # Kakak
])

# Rata rata langkah per orang
avg_per_person = np.mean(data_langkah, axis=1)
print("Rata-rata langkah per orang:", avg_per_person)

# Rata rata langkah per hari
avg_per_day = np.mean(data_langkah, axis=0)
print("Rata-rata langkah per hari:", avg_per_day)
```

```python
# Data durasi ngoding dalam 4 minggu
month_results = np.array([56, 100, 33, 0, 45, 45, 46, 34, 89, 180, 60, 45, 45, 44, 46, 45, 0, 0, 15, 90, 301, 197, 20, 60, 45, 45, 42, 45])
# cek jumlah elemen dalam month_results
print("Shape dari month_results:", month_results.shape)
# ubah bentuk month_results menjadi 4 minggu x 7 hari
weekly_results = month_results.reshape(4, 7)
print("Weekly results:\n", weekly_results)
# hitung rata-rata durasi ngoding per minggu
avg_weekly = np.mean(weekly_results, axis=1)
print("Rata-rata durasi ngoding per minggu:", avg_weekly)
# hitung rata-rata durasi ngoding per hari
avg_daily = np.mean(weekly_results, axis=0)
print("Rata-rata durasi ngoding per hari:", avg_daily)
# visualisasikan data durasi ngoding per minggu
plt.plot(weekly_results.T)
plt.title("Durasi Ngoding per Minggu")
plt.xlabel("Hari dalam Minggu")
plt.ylabel("Durasi Ngoding (menit)")
plt.legend(["Minggu 1", "Minggu 2", "Minggu 3", "Minggu 4"])
plt.show()
```
