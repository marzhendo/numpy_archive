```python
import numpy as np
```

```python
# Tinggi gedung-gedung tertinggi di dunia dalam satuan feet
tallest_buildings = np.array([2717, 2227, 2073, 1972, 1966, 1819, 1776])
# Konversi tinggi dari feet ke meter
print(f'Tinggi gedung dalam meter:\n{tallest_buildings * 0.3048}')
# Coba operasi arithmetic lainnya
print(f'Tinggi gedung ditambah 100 feet:\n{tallest_buildings + 100}')
print(f'Tinggi gedung dikurangi 50 feet:\n{tallest_buildings - 50}')
print(f'Tinggi gedung dikali 2:\n{tallest_buildings * 2}')
print(f'Tinggi gedung dibagi 3:\n{tallest_buildings / 3}')
print(f'Tinggi gedung pangkat 2:\n{tallest_buildings ** 2}')
print(f'Tinggi gedung modulo 500:\n{tallest_buildings % 500}')
```
