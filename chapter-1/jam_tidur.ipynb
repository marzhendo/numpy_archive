```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sleep_hours = np.array([4, 5, 3, 5, 6, 5, 4])
# rata-rata jam tidur
avg_sleep = np.mean(sleep_hours)
print(f'Average sleep hours: {avg_sleep}')
# selisih jam tidur hari pertama dengan hari terakhir
sleep_diff = sleep_hours[-1] - sleep_hours[0]
print(f'Sleep hours difference between first and last day: {sleep_diff}')
# visualisasi jam tidur
hari = np.arange(1, 8)
plt.plot(hari, sleep_hours, marker='o')
plt.title('Jam Tidur selama Seminggu')
plt.xlabel('Hari ke-')
plt.ylabel('Jam Tidur')
plt.axhline(y=avg_sleep, color='r', linestyle='--', label='Rata-rata Jam Tidur')
plt.grid(True)
plt.legend()
plt.show()
```

```python
gelas_air = np.array([8, 7, 9, 6, 8, 7, 10])
avg_water = np.mean(gelas_air)
print(f'Average water intake: {avg_water}')
water_diff = gelas_air[-1] - gelas_air[0]
print(f'Water intake difference between first and last day: {water_diff}')
# membuat array hari ke 1 sampai hari ke 7
hari = np.arange(1, 8)
# plotting data
plt.plot(hari, gelas_air, marker = 'o')
plt.title('Konsumsi Air Mingguan')
plt.xlabel('Hari ke-')
plt.ylabel('Gelas Air')
plt.grid(True)
plt.show()
```
