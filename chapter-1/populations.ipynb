```python
import numpy as np
```

```python
# Slicing = [start:stop:step]
nyc_populations = np.array([19571216,
19673200,
19854526,
20104710,
19463131,
19544098,
19593849,
19636391,
19657321,
19653431,
19626488])
average_population = np.mean(nyc_populations)
print(f'Population of 2023: {nyc_populations[0]}')
print(f'Population of 2013: {nyc_populations[-1]}')
print(f'Average NYC population from 2023 to 2013: {average_population:.0f}')
# Selisih populasi antara tahun 2023 dan 2013
population_difference = nyc_populations[0] - nyc_populations[-1]
print(f'Population difference between 2023 and 2013: {population_difference}')
# Covid-19 terjadi pada sekitar tahun 2020 - 2022
covid_years = nyc_populations[1:4]
covid_population_average = np.mean(covid_years)
print(f'Average NYC population during COVID-19 years (2020-2022): {covid_population_average:.0f}')
# Mengambil data dari awal loncat 2 tahun
every_two_years = nyc_populations[::2]
print(f'NYC populations every two years from 2023 to 2013: {every_two_years}')
# Membalik urutan array
reversed_populations = nyc_populations[::-1]
print(f'NYC populations from 2013 to 2023: {reversed_populations}')
```
